```python
import streamlit as st
import pandas as pd

df = pd.read_csv(r"C:\Users\kayla\OneDrive\Documents\Readmission System\patient_predictions.csv")
```

```python
st.title("Hospital Readmission Risk System")

st.metric("Total Patients", len(df))
st.metric("High Risk Patients", df["Prediction"].sum())
```

```python
risk_filter = st.slider("Risk Threshold", 0.0, 1.0, 0.35)

filtered = df[df["Probability"] >= risk_filter]

st.dataframe(filtered)
```

```python
patient_id = st.selectbox("Select Patient", df.index)

st.write(df.loc[patient_id])
```

```python
st.write("Recommendations:")
st.write(df.loc[patient_id, "Recommendations"])
```
